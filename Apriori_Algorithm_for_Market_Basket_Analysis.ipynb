**Importing Libraries**

```python
import numpy as np
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

**Reading the Data**

```python
df = pd.read_csv('Retail_Data.csv')
```

**Displaying the Data**

```python
df.head()
```

**One-Hot Encoding**

```python
#Create dummy variables
df1 = pd.get_dummies(df) #The get_dummies function in pandas is used to convert categorical variables into dummy/indicator variables.
df2 = df1.iloc[:,1:] #all rows and all columns starting from the second column
df2.head()
```

**Support:** Refers to the number of transactions of an item/ set of items with respect to all the transactions.

```python
#Rules having atleast 5% support
frequent_items = apriori(df2,min_support=0.05, use_colnames=True)
frequent_items
```

**Confidence:** It is the probability of buying product B given the consumer has bought product A

**Lift:** Its value indicates the level of association between the antecedent and consequents.
* Lift<1 suggest a negative association
* Lift=1 suggests no association
* List>1 suggests a strong association

Lift is the ratio of confidence to support

```python
#Rules with minimum lift of 1
rules = association_rules(frequent_items,metric="lift",min_threshold=1)
rules
```

```python
#filter out rules having support>=2, confidence>=0.6 and support>=0.2
rules[(rules['lift']>=2) & (rules['support']>=0.2) & (rules['confidence']>=0.6)]
```

**Conclusion:**

We extrapolate the first row:
1.   21% of all transactions are bread and butter.
2.   61% consumers buy butter given they buy bread.
3. Since lift>1, there is a strong association between bread and butter, retailers should arrange them next to each other for more sales and higher profits.



Some other metrics:

**Leverage:** Leverage measures the difference between the observed frequency of co-occurrence of two items (X and Y) in transactions and the frequency that would be expected if X and Y were independent of each other. In other words, it quantifies how much the occurrence of X and Y together deviates from what would be expected by chance.

*Leverage(X->Y) = Support(X and Y) - Support(X)Support(Y)*

* A positive leverage value indicates that the occurrence of X and Y together is more frequent than what would be expected by chance, suggesting a positive association.

* A leverage value of 0 suggests that the co-occurrence of X and Y is as expected under independence, meaning there is no association.

* A negative leverage value indicates that the occurrence of X and Y together is less frequent than expected, suggesting a negative association or avoidance.
